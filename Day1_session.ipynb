###Sentiment Analysis using Transformer model

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
classifier([
    "I've been waiting for a Hugginface course my whole life.",
    "I have this so much!"
])
```

### Text generation model

```python
generator = pipeline("text-generation")
generator("I am the god of this world")
```

```python
qa = pipeline("question-answering")
qa(
    question="What do i do?",
    context="Hello, I was in my previous life.",
)
```

### Langugae translator

```python
translator = pipeline("translation_en_to_de")
translator("I LOVE GAMES")
```

### NER models

```python
ner = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
ner("I am the god of this world")
```
